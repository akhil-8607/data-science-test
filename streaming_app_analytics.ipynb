creating new columns for date and week of year

```
video_plays_DF = video_plays_DF.withColumn('date',to_date(video_plays_DF.timestamp))
video_plays_DF = video_plays_DF.withColumn('week_of_year',weekofyear(video_plays_DF.timestamp))
```

```
video_plays_DF.show()
```

checking number of unique users in the video plays dataset

```
from pyspark.sql.functions import countDistinct
unique_user_count=video_plays_DF.select(countDistinct("user_id"))
#counting unique user_id in the data frame
unique_user_count.show()
```

```
daily_users_video=video_plays_DF.groupBy("date").agg(countDistinct("user_id"))
#Grouping by date and counting no of users who played video on that date
daily_users_video=daily_users_video.orderBy("date")
#order by date
daily_users_video= daily_users_video.withColumn("Percentage",(daily_users_video['count(user_id)']*100/unique_user_count.collect()[0][0]))
#Caluclating Percentage from no of users who played video each day and total number of users in the data set
daily_users_video= daily_users_video.withColumn('day_of_month',dayofmonth(daily_users_video.date))
#Extracting day of month into new column
daily_users_video.show(31)
#Display all 31 days percemntages 
```

Considering only the users present in the video plays data for analysis since the time period in which data is generated is less compared to the users data set

```
weekly_users_video=video_plays_DF.groupBy("week_of_year").agg(countDistinct("user_id"))

# Grouping users who played video in each week and counting the number of unique users

weekly_users_video=weekly_users_video.orderBy("week_of_year")

# Ordering according to week of year

weekly_users_video=weekly_users_video.withColumn("Percentage",(weekly_users_video['count(user_id)']*100/unique_user_count.collect()[0][0]))

# Caluclatinf percentage of users for each week from the total number of users present in the dataset

weekly_users_video.show()
```

Day of month vs percentage of users who played a video on that day

week of year  vs percentage of users who played a video on that day

```
ax = weekly_users_video.toPandas().plot(x="week_of_year",y="Percentage")
```

Exploring session data based on the time spent on stream app , updating the sessions data frame with time spent on the app during each session

```
diff_secs_col = col("end_timestamp").cast("long") - col("start_timestamp").cast("long")
# Subtracting start time from end time to get the total time spent in each session
sessions_DF = sessions_DF.withColumn( "session_time", diff_secs_col )
sessions_DF.show(truncate=False)
```

Most used devices by view time

```
top_used_devices_time = sessions_DF.groupBy("device_model").agg(sum('session_time').alias("total_session_time")).orderBy("total_session_time",ascending=False)
# Aggregating the view time for each device model and ordering by total view time
top_used_devices_time.show()
```

19 of the 20 top devices by view time are apple devices andd iphone 7 tops among all devices in view time

Most used devices by user count

```
top_used_devices=sessions_DF.groupBy("device_model").agg(countDistinct("user_id"))
#grouping by device model and counting the number of users for each device model 
top_used_devices.orderBy("count(user_id)",ascending=False).show()
```

Among device model numbers with most users in the given sessions data  , Top 20 devices are dominated by apple devices followed samsung models
